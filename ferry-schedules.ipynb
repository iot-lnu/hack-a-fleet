```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the ferry trips data
ferry_data = pd.read_csv('ferry_trips_data.csv')

# Display the first few rows to understand the structure
ferry_data.head()
```

```python
# Summary of the dataset
ferry_data.info()

# Summary statistics
ferry_data.describe()
```

## Analyze Trip Types

Examine the frequency and types of trips to identify potential scheduling improvements.

```python
# Count the number of each trip type
trip_type_counts = ferry_data['trip_type'].value_counts()

# Plot the counts of trip types
sns.barplot(x=trip_type_counts.index, y=trip_type_counts.values)
plt.title('Trip Types Count')
plt.xlabel('Trip Type')
plt.ylabel('Count')
plt.show()
```

## Analyze Vehicles Left at Terminal

Determine how many vehicles are often left at terminals, indicating need for scheduling adjustments.

```python
# Total vehicles left at the terminal
vehicles_left_total = ferry_data['vehicles_left_at_terminal_outbound'].sum() + ferry_data['vehicles_left_at_terminal_inbound'].sum()

# Average vehicles left at terminal per trip
vehicles_left_avg = ferry_data[['vehicles_left_at_terminal_outbound', 'vehicles_left_at_terminal_inbound']].mean()

print(f"Total vehicles left: {vehicles_left_total}")
print(vehicles_left_avg)
```

### Total Vehicles Left: 44,748
- This is the cumulative number of vehicles that were left at the terminal across all trips during the observed period. It might suggest inefficiencies or mismatches in ferry capacity versus demand.

### Average Vehicles Left at Terminal Per Trip:
- **Outbound: 0.318175**
- **Inbound: 0.326999**

These averages represent the mean number of vehicles left at the terminal for outbound and inbound trips, respectively. Although the numbers are low, the significant total indicates that certain trips might have been significantly overloaded.

## Analyze PCE (Passenger Car Equivalent)

Investigate the use of space (PCE) on ferries.


```python
# Calculate and visualize PCE distributions
pce_outbound = ferry_data['passenger_car_equivalent_outbound']
pce_inbound = ferry_data['passenger_car_equivalent_inbound']

sns.histplot(pce_outbound, color='blue', label='Outbound', kde=True)
sns.histplot(pce_inbound, color='orange', label='Inbound', kde=True)
plt.title('Passenger Car Equivalent Distribution')
plt.xlabel('PCE')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
