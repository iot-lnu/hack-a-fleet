# Fuel consumption analysis

```python
def fuel_consumption_analysis(df):
    consumption_data = df[['route_name', 'ferry_name', 'fuelcons_outbound_l', 'fuelcons_inbound_l']]
    
    # Drop null values
    consumption_data = consumption_data.dropna()

    # Aggregating mean fuel consumption by route_name and ferry_name
    fuel_consumption_by_route_ferry = consumption_data.groupby(['route_name', 'ferry_name']).mean().reset_index()

    fuel_consumption_by_route_ferry_melted = fuel_consumption_by_route_ferry.melt(
        id_vars=['route_name', 'ferry_name'],
        value_vars=['fuelcons_outbound_l', 'fuelcons_inbound_l'],
        var_name='direction',
        value_name='average_fuel_consumption_l'
    )

    plt.figure(figsize=(16, 8))
    sns.barplot(
        data=fuel_consumption_by_route_ferry_melted,
        x='route_name',
        y='average_fuel_consumption_l',
        hue='direction'
    )
    plt.xticks(rotation=45, ha='right')
    plt.title('Average Fuel Consumption by Route and Ferry')
    plt.xlabel('Route Name')
    plt.ylabel('Average Fuel Consumption (L)')
    plt.legend(title='Direction')
    plt.tight_layout()
    plt.show()

```

# Distance travelled analysis

```python
def distance_travelled_analysis(df):
    distance_data = df[['route_name', 'ferry_name', 'distance_outbound_nm', 'distance_inbound_nm']]
    
    # Drop null values
    distance_data = distance_data.dropna()

    # Aggregating mean distance by route_name and ferry_name
    distance_by_route_ferry = distance_data.groupby(['route_name', 'ferry_name']).mean().reset_index()

    distance_by_route_ferry_melted = distance_by_route_ferry.melt(
        id_vars=['route_name', 'ferry_name'],
        value_vars=['distance_outbound_nm', 'distance_inbound_nm'],
        var_name='direction',
        value_name='average_distance_nm'
    )

    plt.figure(figsize=(16, 8))
    sns.barplot(
        data=distance_by_route_ferry_melted,
        x='route_name',
        y='average_distance_nm',
        hue='direction'
    )
    plt.xticks(rotation=45, ha='right')
    plt.title('Average Distance Travelled by Route and Ferry')
    plt.xlabel('Route Name')
    plt.ylabel('Average Distance (NM)')
    plt.legend(title='Direction')
    plt.tight_layout()
    plt.show()

# Call the function
distance_travelled_analysis(ferry_data)
```

```python
# Define all the analysis functions as before but for filtered data
def demand_analysis(df):
    demand_by_hour_route = df.groupby(['route_name', 'hour']).agg({
        'vehicles_left_at_terminal_outbound': 'sum',
        'vehicles_left_at_terminal_inbound': 'sum'
    }).reset_index()
    
    demand_by_hour_route['total_vehicles_left'] = (
        demand_by_hour_route['vehicles_left_at_terminal_outbound'] +
        demand_by_hour_route['vehicles_left_at_terminal_inbound']
    )
    
    plt.figure(figsize=(14, 6))
    sns.lineplot(data=demand_by_hour_route, x='hour', y='total_vehicles_left', hue='route_name')
    plt.title('Vehicles Left at Terminal by Hour and Route for Ljusteröleden')
    plt.xlabel('Hour of the Day')
    plt.ylabel('Total Vehicles Left')
    plt.legend(title='Route Name')
    plt.show()

```

```python
def efficiency_improvements(df):
    # Fill NaN values with zeros or appropriate values
    df['fuelcons_outbound_l'] = df['fuelcons_outbound_l'].fillna(0)
    df['fuelcons_inbound_l'] = df['fuelcons_inbound_l'].fillna(0)
    
    # Group and aggregate with more comprehensive statistics
    conditions = df.groupby(['route_name', 'date', 'hour', 'ferry_name']).agg({
        'vehicles_left_at_terminal_outbound': 'sum',
        'vehicles_left_at_terminal_inbound': 'sum',
        'fuelcons_outbound_l': ['mean', 'sum', 'std'],
        'fuelcons_inbound_l': ['mean', 'sum', 'std'],
        'distance_outbound_nm': ['mean', 'sum'],
        'distance_inbound_nm': ['mean', 'sum']
    }).reset_index()

    conditions.columns = ['_'.join(col).strip('_') for col in conditions.columns.values]

    conditions['total_vehicles_left'] = (
        conditions['vehicles_left_at_terminal_outbound_sum'] + 
        conditions['vehicles_left_at_terminal_inbound_sum']
    )
    
    # Determine high demand threshold
    threshold = conditions['total_vehicles_left'].quantile(0.95)
    high_demand_conditions = conditions[conditions['total_vehicles_left'] > threshold]
    
    # Additional metrics for insights
    high_demand_conditions['fuel_efficiency_outbound'] = high_demand_conditions['distance_outbound_nm_sum'] / high_demand_conditions['fuelcons_outbound_l_sum']
    high_demand_conditions['fuel_efficiency_inbound'] = high_demand_conditions['distance_inbound_nm_sum'] / high_demand_conditions['fuelcons_inbound_l_sum']
    
    # Drop or handle NaN results due to calculations
    high_demand_conditions = high_demand_conditions.fillna(0)
    
    return high_demand_conditions
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import json

# Load schedule data
schedule_df = pd.read_csv(schedule_path)


def load_boat_info_df(json_path, boat_key):
    try:
        # Load the JSON data
        with open(json_path, 'r') as file:
            ferry_data = json.load(file)
        
        # Extract the boat information for the specified key
        boat_info = ferry_data.get(boat_key, {})
        
        # Convert the boat information dictionary to a DataFrame
        boat_info_df = pd.DataFrame([boat_info])
        
        return boat_info_df
    except Exception as e:
        print("An error occurred while loading boat info:", e)
        return pd.DataFrame()

def analyze_operations(schedule_df, boat_info_df):
    # If boat information is available, calculate theoretical metrics
    if not boat_info_df.empty:
        max_speed = boat_info_df.iloc[0].get('max_speed', 1)
        fuel_capacity = boat_info_df.iloc[0].get('fuel_capacity', 0)
        
        schedule_df['Theoretical_Trips'] = (schedule_df['Total_Time'] / 60) * max_speed
        schedule_df['Fuel_Needed'] = schedule_df['Theoretical_Trips'] / fuel_capacity
        schedule_df['Efficiency'] = schedule_df['Fuel_Needed'] / (schedule_df['Total_Time'] + 1)
    else:
        schedule_df['Efficiency'] = 1  # Default if no boat info

    return schedule_df

# Load boat information
boat_info_df = load_boat_info_df('ferries.json', 'jupiter')

# Process schedule data
schedule_df['Timestamp'] = pd.to_datetime(schedule_df['Timestamp'])

# Split route into start and end
schedule_df[['Start', 'End']] = schedule_df['Route'].str.split('_', expand=True)

# Calculate time between schedule points
schedule_df['Previous_Timestamp'] = schedule_df.groupby(['Start', 'End'])['Timestamp'].shift()
schedule_df['Total_Time'] = (schedule_df['Timestamp'] - schedule_df['Previous_Timestamp']).dt.total_seconds() / 60

# Analyze operations and calculate efficiency
analyzed_schedule = analyze_operations(schedule_df, boat_info_df)

# Fit a linear regression model on processed data
lr = LinearRegression()
lr.fit(analyzed_schedule[['Total_Time']], analyzed_schedule['Efficiency'])

# Predict a metric based on Total Time (dummy example)
analyzed_schedule['predicted_metric'] = lr.predict(analyzed_schedule[['Total_Time']])

# Sort the dataframe by the predicted metric and Total Time
optimized_schedule = analyzed_schedule.sort_values(by=['predicted_metric', 'Total_Time'], ascending=[True, False])

# Print the optimized schedule
print(optimized_schedule)
```
